# To train the Object detection Model:

```python
import torch
from ultralytics.yolo.utils.checks import check_requirements

# Trust this class
torch.serialization.add_safe_globals([torch.nn.Module])

# Then reload
from ultralytics import YOLO
model = YOLO("yolov8n.pt")  # might still fail if repo version is broken

# # Train the model with 2 GPUs
results = model.train(data='data.yaml', epochs=200, imgsz=640) 
```
